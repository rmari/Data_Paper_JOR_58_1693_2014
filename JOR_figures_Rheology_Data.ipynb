## Averaging data

### Repulsion model

```python
##########
# Normal procedure
vf_list_M1 = sorted(list(set([k[1] for k in data_M1])))
avg_data_M1 = dict()
std_data_M1 = dict()
avg_st_M1 = dict()
std_st_M1 = dict()

# excluding list: the data won't be shown in averages plots
excludes = [ (0.0086,0.57), (0.0088,0.57), (0.0092,0.57),(0.0093,0.57), (0.0094,0.57), (0.0095,0.57), (0.0096,0.57), (0.0097,0.57), (0.0098,0.57), (0.0099,0.57), (0.0102,0.57), (0.0106,0.57)]

for k in data_M1:
    if k not in excludes:
        transient = int(0.05*data_M1[k].shape[0]) # remove the transient
        avg_data_M1[k] = data_M1[k][transient:-1].mean(axis=0)
        std_data_M1[k] = data_M1[k][transient:-1].std(axis=0)
        avg_st_M1[k] = stresses_M1[k][transient:-1].mean(axis=0)
        std_st_M1[k] = stresses_M1[k][transient:-1].std(axis=0)
    
###########
# Special treatment for data showing 2 states

# (0.009,0.57): good data for the 2 states
low_split_pt = 200
high_split_pt = 500
k=(0.009,0.57)
transient = int(0.05*data_M1[k].shape[0]) # remove the transient
avg_data_M1.pop(k)
avg_data_M1[k+('high',)] = data_M1[k][transient:-1][data_M1[k][transient:-1, 1]>high_split_pt].mean(axis=0)
avg_data_M1[k+('low',)] = data_M1[k][transient:-1][data_M1[k][transient:-1, 1]<low_split_pt].mean(axis=0)
std_data_M1.pop(k)
std_data_M1[k+('high',)] = data_M1[k][transient:-1][data_M1[k][transient:-1, 1]>high_split_pt].std(axis=0)
std_data_M1[k+('low',)] = data_M1[k][transient:-1][data_M1[k][transient:-1, 1]<low_split_pt].std(axis=0)
avg_st_M1.pop(k)
avg_st_M1[k+('high',)] = stresses_M1[k][transient:-1][stresses_M1[k][transient:-1, 1]>high_split_pt].mean(axis=0)
avg_st_M1[k+('low',)] = stresses_M1[k][transient:-1][stresses_M1[k][transient:-1, 1]<low_split_pt].mean(axis=0)
std_st_M1.pop(k)
std_st_M1[k+('high',)] = stresses_M1[k][transient:-1][stresses_M1[k][transient:-1, 1]>high_split_pt].std(axis=0)
std_st_M1[k+('low',)] = stresses_M1[k][transient:-1][stresses_M1[k][transient:-1, 1]<low_split_pt].std(axis=0)

# (0.01,0.57), (0.015,0.56): there is one big down fluctuation, but too small to provide good data about low viscosity state
high_split_pt = 300
for k in [ (0.01,0.57), (0.015,0.56) ]:
    transient = int(0.05*data_M1[k].shape[0]) # remove the transient
    avg_data_M1.pop(k)
    avg_data_M1[k] = data_M1[k][transient:-1][data_M1[k][transient:-1, 1]>high_split_pt].mean(axis=0) 
    std_data_M1.pop(k)
    std_data_M1[k] = data_M1[k][transient:-1][data_M1[k][transient:-1, 1]>high_split_pt].std(axis=0)
    avg_st_M1.pop(k)
    avg_st_M1[k] = stresses_M1[k][transient:-1][stresses_M1[k][transient:-1, 1]>high_split_pt].mean(axis=0) 
    std_st_M1.pop(k)
    std_st_M1[k] = stresses_M1[k][transient:-1][stresses_M1[k][transient:-1, 1]>high_split_pt].std(axis=0) 

low_split_pt = 100
# (0.014,0.57): there is one big up fluctuation, but too small to provide good data about low viscosity state
for k in [ (0.014,0.56) ]:
    transient = int(0.05*data_M1[k].shape[0]) # remove the transient
    avg_data_M1.pop(k)
    avg_data_M1[k] = data_M1[k][transient:-1][data_M1[k][transient:-1, 1]<low_split_pt].mean(axis=0) 
    std_data_M1.pop(k)
    std_data_M1[k] = data_M1[k][transient:-1][data_M1[k][transient:-1, 1]<low_split_pt].std(axis=0)
    avg_st_M1.pop(k)
    avg_st_M1[k] = stresses_M1[k][transient:-1][stresses_M1[k][transient:-1, 1]<low_split_pt].mean(axis=0) 
    std_st_M1.pop(k)
    std_st_M1[k] = stresses_M1[k][transient:-1][stresses_M1[k][transient:-1, 1]<low_split_pt].std(axis=0) 

```

```python
##########
# Normal procedure
vf_list_M2_L = sorted(list(set([k[1] for k in data_M2_L])))
avg_data_M2_L = dict()
std_data_M2_L = dict()
avg_st_M2_L = dict()
std_st_M2_L = dict()

# excluding list: the data won't be shown in averages plots
#excludes = [ (0.0086,0.57), (0.0088,0.57), (0.0092,0.57),(0.0093,0.57), (0.0094,0.57), (0.0095,0.57), (0.0096,0.57), (0.0097,0.57), (0.0098,0.57), (0.0099,0.57), (0.0102,0.57), (0.0106,0.57)]

for k in data_M2_L:
    if k not in excludes:
        transient = int(0.05*data_M2_L[k].shape[0]) # remove the transient
        avg_data_M2_L[k] = data_M2_L[k][transient:-1].mean(axis=0)
        std_data_M2_L[k] = data_M2_L[k][transient:-1].std(axis=0)
        avg_st_M2_L[k] = stresses_M2_L[k][transient:-1].mean(axis=0)
        std_st_M2_L[k] = stresses_M2_L[k][transient:-1].std(axis=0)
    
###########
# Special treatment for data showing 2 states

# (0.018,0.56): good data for the 2 states
k=(0.018,0.56)
transient = int(0.05*data_M2_L[k].shape[0]) # remove the transient
split_pt = 200
avg_data_M2_L.pop(k)
avg_data_M2_L[k+('high',)] = data_M2_L[k][transient:-1][data_M2_L[k][transient:-1, 1]>split_pt].mean(axis=0)
avg_data_M2_L[k+('low',)] = data_M2_L[k][transient:-1][data_M2_L[k][transient:-1, 1]<split_pt].mean(axis=0)
std_data_M2_L.pop(k)
std_data_M2_L[k+('high',)] = data_M2_L[k][transient:-1][data_M2_L[k][transient:-1, 1]>split_pt].std(axis=0)
std_data_M2_L[k+('low',)] = data_M2_L[k][transient:-1][data_M2_L[k][transient:-1, 1]<split_pt].std(axis=0)
avg_st_M2_L.pop(k)
avg_st_M2_L[k+('high',)] = stresses_M2_L[k][transient:-1][stresses_M2_L[k][transient:-1, 1]>split_pt].mean(axis=0)
avg_st_M2_L[k+('low',)] = stresses_M2_L[k][transient:-1][stresses_M2_L[k][transient:-1, 1]<split_pt].mean(axis=0)
std_st_M2_L.pop(k)
std_st_M2_L[k+('high',)] = stresses_M2_L[k][transient:-1][stresses_M2_L[k][transient:-1, 1]>split_pt].std(axis=0)
std_st_M2_L[k+('low',)] = stresses_M2_L[k][transient:-1][stresses_M2_L[k][transient:-1, 1]<split_pt].std(axis=0)

# (0.02,0.56) there are some big down fluctuation, but too small to provide good data about low viscosity state
k=(0.02,0.56)
transient = int(0.05*data_M2_L[k].shape[0]) # remove the transient
avg_data_M2_L.pop(k)
avg_data_M2_L[k] = data_M2_L[k][transient:-1][data_M2_L[k][transient:-1, 1]>split_pt].mean(axis=0) 
std_data_M2_L.pop(k)
std_data_M2_L[k] = data_M2_L[k][transient:-1][data_M2_L[k][transient:-1, 1]>split_pt].std(axis=0)
avg_st_M2_L.pop(k)
avg_st_M2_L[k] = stresses_M2_L[k][transient:-1][stresses_M2_L[k][transient:-1, 1]>split_pt].mean(axis=0) 
std_st_M2_L.pop(k)
std_st_M2_L[k] = stresses_M2_L[k][transient:-1][stresses_M2_L[k][transient:-1, 1]>split_pt].std(axis=0) 



vf_list_M2_mu02 = sorted(list(set([k[1] for k in data_M2_mu02])))
avg_data_M2_mu02 = dict()
std_data_M2_mu02 = dict()

for k in data_M2_mu02:
    if k not in excludes:
        transient = int(0.05*data_M2_mu02[k].shape[0]) # remove the transient
        avg_data_M2_mu02[k] = data_M2_mu02[k][transient:-1].mean(axis=0)
        std_data_M2_mu02[k] = data_M2_mu02[k][transient:-1].std(axis=0)
```

```python
def xaxis_top_arrow(axis, xpos, label):

    arrow_body_length = 0.06*(axis.get_ylim()[1]-axis.get_ylim()[0])
    arrow_head_length = 0.8*arrow_body_length
    arrow_bottom = axis.get_ylim()[1] - 1.2*(arrow_body_length + arrow_head_length)
    arrow_head_width = log10(axis.get_xlim()[1]-axis.get_xlim()[0])*log10(xpos)*0.065

    ar = arrow(xpos, arrow_bottom, 0, arrow_body_length, head_width=arrow_head_width, head_length=arrow_head_length, color='k')
    txt = text(0.9*xpos, 0.8*(axis.get_ylim()[1]-axis.get_ylim()[0])+axis.get_ylim()[0], label, fontsize=20, color='k')

    return ar, txt
```

```python
def rheology_plot(ax, theset, ind_ydata, xlim, ylim,  ylab, thedat='avg_data', err=False, xdata='rate', ydata_norm='rate', inv=False, ysc='linear', leg=True, lst='--'):

    if theset == 'M1':
        if thedat == 'avg_data':
            avg_data = avg_data_M1
            std_data =  std_data_M1
        if thedat == 'avg_st':
            avg_data = avg_st_M1
            std_data =  std_st_M1
        vf_list = vf_list_M1[:-3]
        linestyles = linestyles_M1

    if theset == 'M2':
        if thedat == 'avg_data':
            avg_data = avg_data_M2
            std_data =  std_data_M2
        if thedat == 'avg_st':
            avg_data = avg_st_M2
            std_data =  std_st_M2
        std_data =  std_data_M2
        vf_list = vf_list_M2
        linestyles = linestyles_M2
    if theset == 'M2_L':
        if thedat == 'avg_data':
            avg_data = avg_data_M2_L
            std_data =  std_data_M2_L
        if thedat == 'avg_st':
            avg_data = avg_st_M2_L
            std_data =  std_st_M2_L
        std_data =  std_data_M2_L
        vf_list = vf_list_M2_L
        linestyles = linestyles_M2
        
    xsc='log'
    ax.set_xscale(xsc)
    ax.set_yscale(ysc)

    ax.tick_params(direction='in', pad=8)

    if xdata == 'rate':
        ax.set_xlabel(r'$\dot{\gamma}/\dot{\gamma_0}$')
    if xdata == 'stress':
        ax.set_xlabel(r'$\sigma/(\eta_0 \dot{\gamma}_0)$')
    ax.set_ylabel(ylab)

    sgn=1
    if ydata_norm[0] == '-':
        sgn = -1
        ydata_norm = ydata_norm[1:]
            
    for vf in vf_list:

        ydat = array([ avg_data[k][ind_ydata] for k in avg_data if k[1]==vf ])
        stress = array([ k[0]*avg_data[k][1] for k in avg_data if k[1]==vf ])
        sr = array([ k[0] for k in avg_data if k[1]==vf ])
        visc = array([ avg_data[k][1] for k in avg_data if k[1]==vf ])
        
        # avg_data in the comprehension below is not a bug: ensures same ordering as visc array
        ydat_err = array([ std_data[k][ind_ydata] for k in avg_data if k[1]==vf ])
        
        if xdata == 'rate':
            xdat = sr
        if xdata == 'stress':
            xdat = stress
        
        try:
            norm = array([ avg_data[k][int(ydata_norm)] for k in avg_data if k[1]==vf ])
        except ValueError:
            if ydata_norm == 'rate':
                norm = 1
            if ydata_norm == 'stress':
                norm = stress
            if ydata_norm == 'visc':
                norm = visc

        ydat = sgn*ydat/norm
        if inv:
            ydat = 1./ydat
            
        xy = vstack((xdat, ydat, ydat_err))
        xy = xy[:,xy[0,:].argsort()]

        if theset == 'M2' and vf==0.56 and abs(xy[1,4]-xy[1,3]) > abs(xy[1,5]-xy[1,3]):
            xy[:,[4,5]]=xy[:,[5,4]] # two points at sr=0.018
        if leg:
            ytext = xy[1,-1]
            if ytext < ylim[1]:
                text(xy[0,-1]*(1.2+0.05*log10(xlim[1]-xlim[0])), xy[1,-1], r'$\phi ='+str(vf)+'$', fontsize=16, color=linestyles[vf][0])
        
        if err:
            myplot = errorbar(xy[0,:], xy[1,:], yerr=xy[2,:], fmt=lst+linestyles[vf][1], ms=6, color=linestyles[vf][0])
        else:
            myplot = plot(xy[0,:], xy[1,:], lst+linestyles[vf][1], ms=6, color=linestyles[vf][0])

            
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)

    return myplot
```

## Fig. 5

```python
from scipy.stats import gaussian_kde


# Having a smooth histogram is not very easy in Python so far. There's no built-in function I could find.
# Trick 1: The trick here is to use a Kernel Density Estimation (gaussian_kde), and generate data according to the KDE.
# 
# Another problem is that the data are bimodal and spread over 3 orders of magnitude in stress, so the KDE
# should use a variable bandwidth across the x-axis. 
# This is not implemented yet in numpy's gaussian_kde (the bandwidth is uniform), so:
# Trick 2: split the interval according to low stress and high stress states, with a breakpoint at stress=200
# then apply a KDE independantly on the 2 sets of data, with 2 different bandwidths. 
# Of course, to plot it properly, we should know the weight of 
# each data set, which is evaluated before splitting.

fig = plt.figure(1, figsize=(12, 4.5))

#fig.subplots_adjust(hspace=None)

ax=fig.add_subplot(122)
ax.set_xscale('log')
#yscale('log')
ax.set_ylim(0,4.5)
ax.set_xlim(10,10000)
ax.tick_params(direction='in', pad=8)

ax.set_ylabel(r'$P(\log\eta_{\mathrm{r}})$', labelpad=10)
xlabel(r'$\eta_{\mathrm{r}}$')


for k, color, breakpoint in zip(data_points, colors, breakpoints):
    transient = int(0.1*data_M1[k].shape[0])
    data_k = log10(data_M1[k][transient:-1,1])
    density = gaussian_kde(data_k, bw_method=0.2)
    xs = linspace(1,4, 200)
    hstack((0,xs))
    ys = density(xs) 
    plot(power(10,xs), ys, antialiased=True, linewidth=2, color=color, label='$\dot\gamma/\dot\gamma_0 = '+str(k[0])+'$')
    ax.fill_between(power(10,xs), ys, y2=0.0001, alpha=.2, zorder=5, antialiased=True, color=color)


ax.legend(fontsize=17)

ax=fig.add_subplot(121)
ax.set_yscale('log')
ax.set_ylabel(r'$\eta_{\mathrm{r}}$')
ax.set_xlabel(r'$\gamma$')
ax.set_xlim(0,40)
ax.set_ylim(25,100000)
for k, color, breakpoint in zip(data_points, colors, breakpoints):
    visc = data_M1[k][:,1]
    strain = data_M1[k][:,0]
    plot(strain, visc, color=color, label='$\dot\gamma/\dot\gamma_0 = '+str(k[0])+'$', lw=2)

ax.legend(loc=1, fontsize=17)
tight_layout(pad=1.9)
```
